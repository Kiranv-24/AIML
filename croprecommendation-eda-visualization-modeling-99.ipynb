```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split

from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
import xgboost
from xgboost import XGBClassifier

from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
import pickle

import warnings
warnings.filterwarnings("ignore")

%matplotlib inline
```

```python
DATA_PATH = "Crop_recommendation.csv"
```

### Attributes Description

1. N - ratio of Nitrogen content in soil
2. P - ratio of Phosphorous content in soil
3. K - ratio of Potassium content in soil
4. temperature - temperature in degree Celsius
5. humidity - relative humidity in %
6. ph - ph value of the soil
7. rainfall - rainfall in mm

From the paiplot we can see that different clusters are formed based on the feautures in the dataset. First let us experiment with KNearestNeighbors model.

```python
label_encoder = LabelEncoder()
X = data[all_columns]
y = label_encoder.fit_transform(data["label"])
print(X.shape, y.shape)
```

```python
label_dict = {}
for i in range(22):
    label_dict[i] = label_encoder.inverse_transform([i])[0]
label_dict
```

```python
X_train, X_test, y_train, y_test = train_test_split(X.values, y, test_size = 0.2, random_state = 0)
print(f"Train Data: {X_train.shape}, {y_train.shape}")
print(f"Train Data: {X_test.shape}, {y_test.shape}")
```

## KNN Classifier

```python
error_rate = []
for i in range(1, 50):
    pipeline = make_pipeline(StandardScaler(), KNeighborsClassifier(n_neighbors = i))
    pipeline.fit(X_train, y_train)
    predictions = pipeline.predict(X_test)
    accuracy = accuracy_score(y_test, predictions)
    print(f"Accuracy at k = {i} is {accuracy}")
    error_rate.append(np.mean(predictions != y_test))
print(X_train.shape)  # Check if it has 8 features
print(X_test.shape)   # Check if it has 8 features
print(X.shape)        # Check if it has 8 features

plt.figure(figsize=(10,6))
plt.plot(range(1,50),error_rate,color='blue', linestyle='dashed', 
         marker='o',markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
print("Minimum error:-",min(error_rate),"at K =",error_rate.index(min(error_rate))+1)
```

```python
# Optimal K value is 4
knn_pipeline = make_pipeline(StandardScaler(), KNeighborsClassifier(n_neighbors=4))

# Fit the model with 7 features
knn_pipeline.fit(X_train, y_train)

# Check if the data has 7 features (adjust for your dataset accordingly)
print("Shape of X_train:", X_train.shape)  # Check if it has 7 features
print("Shape of X_test:", X_test.shape)    # Check if it has 7 features
print("Shape of X:", X.shape)              # Check if it has 7 features

# Test Data Metrics
predictions = knn_pipeline.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy on Test Data: {accuracy*100}%")

# Confusion Matrix for Test Data
plt.figure(figsize=(15, 9))
sns.heatmap(confusion_matrix(y_test, predictions), annot=True)
plt.title("Confusion Matrix for Test Data")
plt.show()

print()

# Whole Data Metrics
predictions = knn_pipeline.predict(X.values)
accuracy = accuracy_score(y, predictions)
print(f"Accuracy on Whole Data: {accuracy*100}%")

# Confusion Matrix for Whole Data
plt.figure(figsize=(15, 9))
sns.heatmap(confusion_matrix(y, predictions), annot=True)
plt.title("Confusion Matrix for Whole Data")
plt.show()
```

## Random Forest Classifier

```python
# Random Forest Pipeline with 7 features
rf_pipeline = make_pipeline(StandardScaler(), RandomForestClassifier(random_state=18))

# Fit the model with 7 features
rf_pipeline.fit(X_train, y_train)

# Check if the data has 7 features
print("Shape of X_train:", X_train.shape)  # Check if it has 7 features
print("Shape of X_test:", X_test.shape)    # Check if it has 7 features
print("Shape of X:", X.shape)              # Check if it has 7 features

# Accuracy on Test Data
predictions = rf_pipeline.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy on Test Data: {accuracy*100}%")

# Confusion Matrix for Test Data
plt.figure(figsize=(15, 9))
sns.heatmap(confusion_matrix(y_test, predictions), annot=True)
plt.title("Confusion Matrix for Test Data")
plt.show()

print()

# Accuracy on Whole Data
predictions = rf_pipeline.predict(X.values)
accuracy = accuracy_score(y, predictions)
print(f"Accuracy on Whole Data: {accuracy*100}%")

# Confusion Matrix for Whole Data
plt.figure(figsize=(15, 9))
sns.heatmap(confusion_matrix(y, predictions), annot=True)
plt.title("Confusion Matrix for Whole Data")
plt.show()
```

## XGBoost Classifier

```python
# XGBoost Pipeline with 7 features
xgb_pipeline = make_pipeline(StandardScaler(), XGBClassifier(random_state=18))

# Fit the model with 7 features
xgb_pipeline.fit(X_train, y_train)

# Check if the data has 7 features
print("Shape of X_train:", X_train.shape)  # Check if it has 7 features
print("Shape of X_test:", X_test.shape)    # Check if it has 7 features
print("Shape of X:", X.shape)              # Check if it has 7 features

# Accuracy on Test Data
predictions = xgb_pipeline.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy on Test Data: {accuracy*100}%")

# Confusion Matrix for Test Data
plt.figure(figsize=(15, 9))
sns.heatmap(confusion_matrix(y_test, predictions), annot=True)
plt.title("Confusion Matrix for Test Data")
plt.show()

print()

# Accuracy on Whole Data
predictions = xgb_pipeline.predict(X.values)
accuracy = accuracy_score(y, predictions)
print(f"Accuracy on Whole Data: {accuracy*100}%")

# Confusion Matrix for Whole Data
plt.figure(figsize=(15, 9))
sns.heatmap(confusion_matrix(y, predictions), annot=True)
plt.title("Confusion Matrix for Whole Data")
plt.show()
```

```python
import pickle
import numpy as np

# Load the saved models and label dictionary
knn_pipeline = pickle.load(open("knn_pipeline.pkl", "rb"))
rf_pipeline = pickle.load(open("rf_pipeline.pkl", "rb"))
xgb_pipeline = pickle.load(open("xgb_pipeline.pkl", "rb"))
label_dict = pickle.load(open("label_dictionary.pkl", "rb"))

# Example input data (matching feature columns)
input_data = {
    "N": 90,
    "P": 42,
    "K": 43,
    "temperature": 20.879744,
    "humidity": 82.002744,
    "ph": 6.502985,
    "rainfall": 202.935536
}

# Prepare the input data for prediction
column_names = ["N", "P", "K", "temperature", "humidity", "ph", "rainfall"]
input_array = np.asarray([float(input_data[i]) for i in column_names]).reshape(1, -1)

# Make predictions using each model
xgb_pred = label_dict[xgb_pipeline.predict(input_array)[0]]
rf_pred = label_dict[rf_pipeline.predict(input_array)[0]]
knn_pred = label_dict[knn_pipeline.predict(input_array)[0]]

# Print the results
print("XGBoost Prediction:", xgb_pred)
print("Random Forest Prediction:", rf_pred)
print("KNN Prediction:", knn_pred)

# Combine predictions from all models
predictions = [xgb_pred, rf_pred, knn_pred]
if len(set(predictions)) == len(predictions):
    final_prediction = max(set(predictions), key=predictions.count)
else:
    final_prediction = max(set(predictions), key=predictions.count)

print("Final Prediction:", final_prediction)
```
